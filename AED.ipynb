```python
# Extract clean EDA and SCR features
cleaned = signals["EDA_Clean"]
features = [info["SCR_Onsets"], info["SCR_Peaks"], info["SCR_Recovery"]]
```

```python
Moyenne_SCL  = cleaned.mean()
print("SCL moyen :", Moyenne_SCL)
```
